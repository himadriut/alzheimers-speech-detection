# Notebook 05: Paired Statistical Tests (ASR ablation + classifier comparison)

Requires the score files produced in Notebook 04.

```
import numpy as np
from scipy import stats
from pathlib import Path

PROJECT_ROOT = Path(".").resolve()

def paired_test(a_path: Path, b_path: Path, name: str):
    a = np.loadtxt(a_path)
    b = np.loadtxt(b_path)
    diff = a - b
    t_stat, p_two = stats.ttest_rel(a, b)
    print("\n", name)
    print("-"*len(name))
    print(f"Mean(a): {a.mean():.4f} | Mean(b): {b.mean():.4f}")
    print(f"Mean improvement (a-b): {diff.mean():.4f}")
    print(f"Std(diff): {diff.std(ddof=1):.4f}")
    print(f"t-stat: {t_stat:.4f} | two-sided p: {p_two:.6f} | one-sided p: {p_two/2:.6f}")

# ASR ablation: small vs base
paired_test(PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_small.txt",
            PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_base.txt",
            "ASR effect (LR): small vs base")

paired_test(PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_small_svm.txt",
            PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_base_svm.txt",
            "ASR effect (SVM): small vs base")

# Classifier comparison under Whisper small
paired_test(PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_small_svm.txt",
            PROJECT_ROOT/"outputs"/"repeated_tfidf_scores_small.txt",
            "Classifier effect (Whisper small): SVM vs LR")
```
