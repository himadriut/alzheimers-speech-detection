# ADReSSo Dementia Detection — Notebook 01: Setup & Data Check

This notebook verifies your project structure and confirms the ADReSSo diagnosis training data is in the expected paths.

**Expected structure**
```
adressso_project/
  data/raw/ADReSSo21/diagnosis/train/audio/ad/*.wav
  data/raw/ADReSSo21/diagnosis/train/audio/cn/*.wav
  outputs/
  scripts/
```

> Run this notebook from the repo root: `adressso_project/`

````
from pathlib import Path

PROJECT_ROOT = Path(".").resolve()
DATA_AD = PROJECT_ROOT / "data" / "raw" / "ADReSSo21" / "diagnosis" / "train" / "audio" / "ad"
DATA_CN = PROJECT_ROOT / "data" / "raw" / "ADReSSo21" / "diagnosis" / "train" / "audio" / "cn"

print("Project root:", PROJECT_ROOT)
print("AD path:", DATA_AD)
print("CN path:", DATA_CN)

assert DATA_AD.exists(), f"Missing: {DATA_AD}"
assert DATA_CN.exists(), f"Missing: {DATA_CN}"

ad_files = sorted(DATA_AD.glob("*.wav"))
cn_files = sorted(DATA_CN.glob("*.wav"))

print("AD wav count:", len(ad_files))
print("CN wav count:", len(cn_files))
print("AD sample:", [p.name for p in ad_files[:5]])
print("CN sample:", [p.name for p in cn_files[:5]])
````

If counts look correct (typically **AD=87**, **CN=79**) you’re ready for ASR transcription.
